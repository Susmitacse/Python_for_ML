#MatplotLib
Matplotlib is a plotting library for python programming language and its numerical mathematics extension is NumPy. 

````
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
````

````
x = np.arange(0,10)
y = np.arange(11,21)
````

````
a = np.arange(40,50)
b = np.arange(50,60)
````

````
## Plotting using Matplotlib
## plt scatter

plt.scatter(x,y,c='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph in 2D")
plt.savefig("graph1.jpg")
````

````
y=x*x
````

````
## plt plot

plt.plot(x, y,'r*--', linewidth=2, markersize=10 )
plt.xlabel("X-axis")       # r*-- is red color dashed line with star
plt.ylabel("Y-axis")
plt.title("Graph in 2D")
````

````

plt.plot(x, y,'yo--') # y is for yellow and 'o' is for dotted
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph in 2D")
````

````
## Creating Subplots

plt.subplot(2,2,1)   # No. of rows, columns and position
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*-')
plt.subplot(2,2,4)
plt.plot(x,y,'yo-')
````

````
x = np.arange(1,11)
y = 3 * x + 10
plt.title("Matplotlib Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
````

````
# Compute the x and  y coordinates for points on a sine curve

x = np.arange(0, 4 * np.pi, 0.1)   # 0.1 is the step size
y = np.sin(x)
plt.title('sin curve')
plt.plot(x,y)
````

````
# Compute the x and y coordinates for the points on the sine and cosine   curves

x = np.arange(0, 5 *np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up the subplot grid that has width 2 and  height 1
plt.subplot(2,1,1)
# Make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

# Set the second subplot
plt.subplot(2,1,2)

plt.plot(x, y_cos, 'g--')
plt.title('Cosine')  
````

### Bar Graph

````
# Bar plot
x =[2,8,10]
y =[11,16,9]

x2 =[3,9,11]
y2 =[6,15,7]
plt.bar(x,y, align = 'center')
plt.bar(x2, y2, color='g')
plt.title('Bar graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
````

### Histogram

````
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
````

###Box Plot using Matplotlib

````
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
````

````
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False)
````

###Pie Chart


````
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
````

````
import seaborn as sns
````

### Correlation with Heatmap
A correleation heatmap uses colored cells, typically in a monochromatic scale to show a 2D correlation matrix between two discrete dimensions or event type. It is vey important in feature selection.

````
df.corr()   # correlation find the relation between the numerical attributes. It ranges from -1 to 1
````

````
sns.heatmap(df.corr())
````

### Joint Plot
Join plot helps to study the relation between two numeric variables. The central chart display their correlation. It is usually a scatter plot, a hexbin plot, a 2D histogram or a 2D density plot. 

#### Univariate Analysis

````
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
````

````
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # reg-> regression
````

### Pair Plot 
A pair plot is also known as scatterplot, in which one variable in the same data row is matched with another vriable's value, like this : Pair plots arejust elaborations on this, showing all variables paired with all the other variable. 

````
sns.pairplot(df)
````

````
sns.pairplot(df, hue='sex')  # plots differentaited based on categorical value "sex"
````

````
sns.pairplot(df, hue='smoker')  # plots differentaited based on categorical value "smoker"
````

````
df['smoker'].value_counts()
````

### Dist Plot
Dist plot helps us to check the distribution of columns  feature wise.

````
sns.distplot(df['tip'])
````

````
sns.distplot(df['tip'], kde =False, bins=10)  # kde-> kernel density estimation
````

####Count Plot

#### Bar Plot
In bar plot we have to specify both x and y value, unlike count plot where we specified only x or y value.


````
sns.barplot(x='total_bill', y='sex',data=df) 
````

````
sns.barplot(y='total_bill', x='smoker',data=df) 
````

#### Box plot
A Box plots show the five-number summary of a set of data: including the minimum score, first (lower) quartile, median, third (upper) quartile, and maximum score



````
sns.boxplot('smoker', 'total_bill', data=df)
````

````
sns.boxplot(x='day', y='total_bill', data=df, palette='rainbow')
````

````
sns.boxplot(x='total_bill',y='day', data=df, hue='smoker' )
````

#### Violin Plot
Violin plot helps us to see both the distribution of data in terms of Kernal density estimation and the box plot .

````
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
````
